```python
new_row = np.empty((1,36))
for x in unique_counties:
    group = census_data.loc[census_data['County'] == x]
    for i in census_data.columns:
        if i == 'County' or i == 'State':
            new_row = np.append(new_row, x)
        if census_data[i].dtypes == 'int64':
            new_row = np.append(new_row, sum(group[i]))
        if census_data[i].dtypes == 'float64':
            new_row = np.append(new_row, (sum(group[i])/group[i].size)*100)
    temp = pd.DataFrame(new_row)
    new_census.append(temp)
    new_row = np.empty((1,36))
```
